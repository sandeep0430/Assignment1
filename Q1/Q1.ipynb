```python
import pandas as pd
```

```python
path='/content/Raw.csv'
```

```python
df_1=df.dropna()
```

```python
df_1.to_csv('cleanData.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Count the values for each category
frailty_counts = df_1['Frailty'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(frailty_counts, labels=frailty_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Distribution of Frailty')
plt.show()
```
