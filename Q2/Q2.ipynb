```python
df1=df.dropna()
```

```python
df1.to_csv('cleanData.csv',index=False)
```

```python
df1.info()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df1['math score'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df1, x='gender', y='math score', palette='coolwarm')
plt.title('Math Score Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df1, y='parental level of education', order=df1['parental level of education'].value_counts().index, palette='viridis')
plt.title('Parental Level of Education')
plt.xlabel('Count')
plt.ylabel('Parental Education Level')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df1, x='math score', y='reading score', hue='gender', alpha=0.7, palette='husl')
plt.title('Math vs. Reading Scores')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(data=df1, x='lunch', y='writing score', palette='Set2')
plt.title('Writing Scores by Lunch Type')
plt.xlabel('Lunch Type')
plt.ylabel('Writing Score')
plt.show()
```
